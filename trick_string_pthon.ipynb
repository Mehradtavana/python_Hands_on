```python
s = '''I am human, but I need to practice to be better than before. 
I need to focus on now and have prespective of future, I need to practice for
this and I should be ready for everything in near and long future.'''

t = s.lower()
print(t)

t = s.upper()
print(t)

print(len(s))
print(s.count('for'))
print(s.find('but'))

t = s.split('\n')
print(t)
string = '\t'.join(t)
print(string)
```
