```python
import re
```

```python
#meta charcters
s = 'ac'
re.search('[acd]',s) # [] match specific
re.search('.',s) # match all charcters except \n
re.search('^a',s) # is used to check if a string start with a certain charcter=re.search('c$', s) # check the string end with charcter or not

# * matches zero or more occurences of the pattern left to it (just show first one location)
a = re.search('ma*','men')
b = re.search('ma*','women')
# + matches one or more occurences of the pattern left to it (just show first one location)
a = re.findall('ma+','man')
b = re.search('ma+','woman')
print(a)
print(b)

# {} exact number of occurences
a = re.search('a{2}','abc')
b = re.search('a{2}','aac')
print(a)
print(b)

# vertical bar | => match one of the side pattern to the string (or)
a = re.search('a|b','mumbi')
b = re.search('a|b','cures')
print(a)
print(b)

# () group subpatterns
a = re.search('(a)m','ammm') # any string that a followed by m
b = re.search('(ab)m','abm')
print(a)
print(b)


# \ use to ignore all meta charcter and see them as normal charcter but should follow with next charcter
b = re.search('\$a','$adg')
print(b)
```

#built-in functions for regular expressions
1. findall() -> return all matches in list :(
2. search()
3. split() -> return a list that string is split in each match
re.split('\s',string) # split based on space
4. sub() -> replaces the matches with the text of our choice in a string
re.sub('\s','2',string) # replace all spaces with 2

#Special Characters for regular expressions
1. \A returns a match if the specified characters are t begining of the string
2. \b returns a match where the specified characters are at the beginning or at the end of a word
3. \B returns a match where the specified characters are present but not at the beginning or at the end of a word
4. \D returns a match where the string does not contain digits
5. \d returns a match where the string contain digits
6. \s return white space characters


```python
s = "I love watching12 movies and I love2 action movies24"
x = re.findall('\AI',s) #give the first match
print(x)

x = re.findall(r'\bmov',s) #r show at beginning
print(x)

x = re.findall('\Bov',s)
print(x)

x = re.findall('\D',s)
print(x)

x = re.findall('\d',s)
print(x)
```
