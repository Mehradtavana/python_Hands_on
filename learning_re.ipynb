```python
import re
```

```python
#meta charcters
s = 'ac'
re.search('[acd]',s) # [] match specific
re.search('.',s) # match all charcters except \n
re.search('^a',s) # is used to check if a string start with a certain charcter=
re.search('c$', s) # check the string end with charcter or not

# * matches zero or more occurences of the pattern left to it
a = re.search('ma*','men')
b = re.search('ma*','women')
# + matches one or more occurences of the pattern left to it
a = re.findall('ma+','man')
b = re.search('ma+','woman')
print(a)
print(b)

# {} exact number of occurences
a = re.search('a{2}','abc')
b = re.search('a{2}','aac')
print(a)
print(b)

# vertical bar | => match one of the side pattern to the string (or)
a = re.search('a|b','mumbi')
b = re.search('a|b','cures')
print(a)
print(b)

# () group subpatterns
a = re.search('(a)m','ammm') # any string that a followed by m
b = re.search('(ab)m','abm')
print(a)
print(b)


# \ use to ignore all meta charcter and see them as normal charcter but should follow with next charcter
b = re.search('\$a','$adg')
print(b)
```

#Special Characters for regular expressions
1. \A returns a match if the specified characters are t begining of the string
2. \b returns a match where the specified characters are at the beginning or at the end of a word
r"\bain"
r"ain\b"
3. \B returns a match where the specified characters are present but not at the beginning or at the end of a word
r"\Bain"
r"ain\B"
4. \D returns a match where the string does not contain digits
5. \d returns a match where the string contain digits
6. \s return white space characters
7. \w Returns a match where the string contains any word characters (characters from a to Z, digits from 0-9, and the underscore _ character)
8. \W Returns a match where the string DOES NOT contain any word characters
9. \Z Returns a match if the specified characters are at the end of the string

```python
s = "I love watching12 movies and I love2 action movies24"
x = re.findall('\AI',s) #give the first match
print(x)

x = re.findall(r'\bmov',s) #r show at beginning
print(x)

x = re.findall('\Bov',s)
print(x)

x = re.findall('\D',s)
print(x)

x = re.findall('\d',s)
print(x)
```

# Sets of regular expressions
A set is a set of characters inside a pair of square brackets [] with a special meaning
1. [arn] Returns a match where one of the specified characters (a, r, or n) are present
2. [a-n] Returns a match for any lower case character, alphabetically between a and n
3. [^arn] Returns a match for any character EXCEPT a, r, and n
4. [0123] Returns a match where any of the specified digits (0, 1, 2, or 3) are present
5. [0-9] Returns a match for any digit between 0 and 9
6. [0-5][0-9] Returns a match for any two-digit numbers from 00 and 59
7. [a-zA-Z] Returns a match for any character alphabetically between a and z, lower case OR upper case
8. [+] In sets, +, *, ., |, (), $,{} has no special meaning, so [+] means: return a match for any + character in the string

```python
s = "I love watching12 movies"
x = re.findall('[lwa]',s)
print(x)

x = re.findall('[a-f]',s)
print(x)

x = re.findall('[^lwa]',s)
print(x)

x = re.findall('[0-9]',s)
print(x)

x = re.findall('[a-zA-Z]',s)
print(x)

x = re.findall('[a-z][0-2]',s)
print(x)
```

# Match Object
A Match Object is an object containing information about the search and the result.
1. .span() returns a tuple containing the start-, and end positions of the match.
2. .string returns the string passed into the function
3. .group() returns the part of the string where there was a match

```python
txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.span()) 
print(x.string)
print(x.group())
```
